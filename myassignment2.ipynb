1. Write a program to determine if a triangle is equilateral, isosceles, or scalene based on user input sides.

```python
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))

if a == b == c:
    print("The triangle is Equilateral.")
elif a == b or b == c or a == c:
    print("The triangle is Isosceles.")
else:
    print("The triangle is Scalene.")
```

2. Input marks of 5 subjects and decide whether the student passed (must score >33 in all) and their average percentage.

```python
marks = []
for i in range(1, 6):
    mark = int(input(f"Enter marks for subject {i}: "))
    marks.append(mark)

if all(mark > 33 for mark in marks):
    print("The student has passed.")
else:
    print("The student has failed.")

average = sum(marks) / len(marks)
print(f"Average Percentage: {average}%")
```

3. Count the frequency of characters in a string using a dictionary.



```python
text = input("Enter a string: ")


freq = {}
for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print("Character Frequencies:", freq)
```

4. Convert two lists into a dictionary (one list of keys, one list of values).



```python
keys = ["name", "age", "city"]
values = ["Ali", 20, "Lahore"]

my_dict = {}

for i in range(len(keys)):
    my_dict[keys[i]] = values[i]

print("Resulting dictionary:", my_dict)
```

5. Try converting a string to an integer and catch any value error if it fails.



```python
num_str = input("Enter an integer: ")

try:
    number = int(num_str)
    print("Conversion successful:", number)
except ValueError:
    print("Error: Please enter a valid integer.")
```

6. Handle a key error while accessing a non-existent key from a dictionary.



```python
my_dict = {"name": "ahmed", "age": 25}

try:
    print(my_dict["work"])
except KeyError:
    print("Error: Key does not exist in the dictionary.")
```

7. Use a try-except-finally block to open a file, read its contents, and close the file (even if an error occurs).



```python
try:
    file = open("default.txt", "r")
    content = file.read()
    print("File Content:\n", content)
except FileNotFoundError:
    print("Error: File not found.")
finally:
    try:
        file.close()
        print("File closed.")
    except:
        pass
```

8. Nest multiple try-except blocks to handle file read and data conversion.



```python
try:
    file = open("data.txt", "r")
    try:
        data = file.read()
        number = int(data)
        print("Number from file:", number)
    except ValueError:
        print("Error: Could not convert data to integer.")
    finally:
        file.close()
        print("File closed.")
except FileNotFoundError:
    print("Error: File not found.")
```
