1. Create a list of all numbers between 1 and 100 that are divisible by both 3 and 5.

```python
list_1=[]
for num in range(100):
    num=num+1
    list_1.append(num)

divisible_by_3_n_5=[]
for num in list_1:
    if num % 3 ==0 and num % 5==0 :
        divisible_by_3_n_5.append(num)
print("The numbers which are divisible by both 3 & 5 are:" , divisible_by_3_n_5)
```

2. Write a loop that replaces all spaces in a string with underscores.

```python
word="This is my bike"
word_with_=""

for i in word:
    if i==" ":
        word_+="_"
    else:
        word_+= i    

        
print(word_with_)
```

3. Check which elements are common in two lists

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

common = []

for num in list1:
    if num in list2:
        common.append(num)

print("Common elements:", common)
```

1. Keep removing elements from a list one by one and print the list each time, until it becomes empty

```python
my_list = [1, 2, 3, 4, 5]
index = len(my_list)

while index > 0:
    index = index - 1
    my_list = my_list[:index] 
    print(my_list)
```

2. Start from a number and keep dividing it by 2 until it becomes less than 1. Print each value.


```python
num1 = 50

while num1 >= 1:
    print(num1)
    num1 = num1 / 2
```

3. Ask the user to enter numbers until the total sum crosses 100.

```python
total = 0

while total <= 100:
    number = int(input("Enter number: "))
    total += number
    print("Current total :", total)

print("Total crossed 100")
```
